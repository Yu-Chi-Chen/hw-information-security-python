### def decimalToBinary(num:int) input: int  output: string

```python
import math

def decimalToBinary(num:int):
    divisor=0
    arrList=list()
    arrList.clear()

    dividend=num
    ## 沒有寫floor會出錯
    while(dividend/2!=0):
            divisor+=1
            dividend=math.floor(dividend/2)
    ## end while

    while(divisor>=0):
        if(num>=math.pow(2,divisor)):
                num-=math.pow(2,divisor)
                divisor-=1
                arrList.append('1')
        else:
                divisor-=1
                arrList.append('0')
            ##end if-else
    ## end while


    while(len(arrList)<8):
        arrList.insert(0,0)
    ## end while

    binary_string: str = ""
    for char_num in arrList:
        binary_string = binary_string + str(char_num)

    # str: str="".join(arrList)
    # if __name__=='_main_':
    #     return str
    
    return binary_string
##end define
```

### def readIMGtopixel(path) input: image path  output: imgpixelmatrix

```python
from PIL import Image
import numpy as np

WIDTH=int(0)
HEIGHT=int(0)
def readIMGtopixel(path):
    img: Image =Image.open(path)
    imgMatrix = np.array(Image.open(path).convert('L'),'i') #dtype=int
    HEIGHT=int(img.height)
    WIDTH=int(img.width)
    return imgMatrix,HEIGHT,WIDTH
#end def
```

### def binaryToDecimal(num) input: int32 output: decimal

```python
def binaryToDecimal(num):
    temp=num
    dec=int(0)
    for i in range (len(str(num))):
        if((temp%10)==1):
            dec+=math.pow(2,i)
        
        temp=int(temp/10)
    #end for
    return dec;
```

```python
def readTextFile(path):
    with open(path,'r') as f:
        text=f.read()

    return text
```

```python
with open("..\Txt_File .txt","r") as f:
    txtContent=f.read()
    print(txtContent)
```

##### ASCII shift 4

```python
outputTXT="";
originASCII=[]
#convert to binary
for element in txtContent:
    originASCII.append(ord(element))
#end for
print(originASCII)

# shift 4(+4)
shiftASCII=[]
shiftText=""
for i in range(len(originASCII)):
    num=int(originASCII[i])+4
    if(num>255):
        num=num-256
    
    shiftASCII.append(num)
    shiftText+=chr(num)
#end for

print(shiftASCII)
print(shiftText)
writeTextWithString(".\shiftText.txt",shiftText)
```

### covert to binary

```python
temp=""
for i in range(len(shiftASCII)):
    temp+=str(decimalToBinary(shiftASCII[i]))

print(temp)
writeTextWithString(".\shiftBinary.txt",temp)
```

### 將shift回來的數字轉成ASCII

```python
shiftBackText=""
for i in range(0,len(shiftBackDec)):
    shiftBackText+=chr(shiftBackDec[i])

print(shiftBackText)
```

## 圖檔部分

```python
imgMatrix,HEIGHT,WIDTH=readIMGtopixel('..\Img_File.bmp')
print(imgMatrix)
shiftIMGMatrix=np.zeros(imgMatrix.shape,dtype=int)
for i_vertical in range(HEIGHT):
    for i_horizon in range(WIDTH):
        num=imgMatrix[i_vertical][i_horizon]+4
        if(num>255):
            num-=256
        shiftIMGMatrix[i_vertical][i_horizon]=num

print(shiftIMGMatrix)
```

### convert shift matrix into a picture

```python
shiftIMG=Image.fromarray(shiftIMGMatrix)
shiftIMG=shiftIMG.convert('L')
shiftIMG.save(".\AfterShiftIMG.bmp")
shiftIMG.show()
```
